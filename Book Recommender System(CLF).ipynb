# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index(name='avg_rating')
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

# Collaborative Filtering Based Recommender System

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.groupby('User-ID').count()
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
x[x]
#these users have done rating on more than 200 books
```

```python
well_studied_users = x[x].index
```

```python
ratings_with_name['User-ID'].isin(well_studied_users)
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(well_studied_users)]
# filtered on the basis of users
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
y[y]
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
filtered_rating['Book-Title'].isin(famous_books)
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
#famous books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores[0]
#this is the similarity of 1984 with every book
```

```python
similarity_scores.shape
```

```python
list(enumerate(similarity_scores[0]))
```

```python
sorted(list(enumerate(similarity_scores[0])), key=lambda x:x[1], reverse=True)
#sorting on the basis of similarity scores
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:11]
    
    data=[]
    
    for i in similar_items:
        item=[]
        print(pt.index[i[0]])
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Hobbit : The Enchanting Prelude to The Lord of the Rings')
```
